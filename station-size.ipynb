```python
import pandas as pd
from plotly.graph_objects import Scatter, Figure
```

```python
voice = pd.read_csv('voice.csv')
geo = pd.read_csv('geo-categorized.csv')
df = geo.merge(voice, on='id', how='left')
```

```python
df[~df['officialVoters'].isna()]
```

```python
areas = ['city', 'minsk_suburb', 'capital', 'village', 'town_below100', 'town_over100', 'embassy']
```

```python
fig = Figure(layout_title='Polling station size')

for area in areas:
    data = df[df['area'] == area]



    fig.add_trace(Scatter(
        x=df.index, 
        y=data['officialVoters'],
        mode='markers',
        name=area,
        text=data['id'],
    ))

fig.update_xaxes(title='Poll stations')
fig.update_yaxes(title='Size')


fig.show()
```

```python
size_data = []
outliers = [
    '07-003-0011', '07-004-0022', '07-005-0002', '07-008-0070', '07-002-0012',
]

for area in areas:
    data = df[(df['area'] == area) & ~df['id'].isin(outliers)]
    avg = data['officialVoters'].mean()
    std = data['officialVoters'].std()
    size_data.append({
        'area': area,
        'size': avg,
        'std': std,
        'std/mean': std / avg,
    })
```

```python
pd.DataFrame(size_data).to_csv('poll-station-size.csv', index=False, float_format='%.3f')
```
